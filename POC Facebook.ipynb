```python
# Obtener datos de campaña
url = f'https://graph.facebook.com/v18.0/act_{ad_account_id}/campaigns'
params = {
    'access_token': access_token,
    'fields': (
        'id,name,status,'
        'insights.time_increment(1){'
        'date_start,impressions,clicks,unique_clicks,cpc,cpm,ctr,spend,frequency,reach}'
    ),
    'summary': 'true'
}

response = requests.get(url, params=params)
if response.status_code == 200:
    campaigns_data = response.json()['data']
    insights_list = []
    for campaign in campaigns_data:
        insights = campaign.get('insights', {}).get('data', [])
        for insight in insights:
            insight_data = {
                'date': insight['date_start'],
                'campaign_id': campaign['id'],
                'campaign_name': campaign['name'],
                'campaign_status': campaign['status'],
                **{key: insight[key] for key in insight if key != 'date_start'}
            }
            insights_list.append(insight_data)
    df_campaign = pd.DataFrame(insights_list)
    df_campaign.set_index('date', inplace=True)
    print(df_campaign)
else:
    print(f'Error: {response.status_code}')
    print(response.text)
```

```python
# Obtener datos de segmentaciones (adsets)
url_adsets = f'https://graph.facebook.com/v18.0/act_{ad_account_id}/adsets'
params_adsets = {
    'access_token': access_token,
    'fields': (
        'id,name,status,'
        'insights.time_increment(1){'
        'date_start,impressions,clicks,unique_clicks,cpc,cpm,ctr,spend,frequency,reach}'
    ),
    'summary': 'true'
}

response_adsets = requests.get(url_adsets, params=params_adsets)
if response_adsets.status_code == 200:
    adsets_data = response_adsets.json()['data']
    insights_list_adsets = []
    for adset in adsets_data:
        insights_adsets = adset.get('insights', {}).get('data', [])
        for insight in insights_adsets:
            insight_data_adsets = {
                'date': insight['date_start'],
                'adset_id': adset['id'],
                'adset_name': adset['name'],
                'adset_status': adset['status'],
                **{key: insight[key] for key in insight if key != 'date_start'}
            }
            insights_list_adsets.append(insight_data_adsets)
    df_adsets = pd.DataFrame(insights_list_adsets)
    df_adsets.set_index('date', inplace=True)
    print(df_adsets)
else:
    print(f'Error: {response_adsets.status_code}')
    print(response_adsets.text)
```

```python
# Obtener datos de anuncios (ads)
url_ads = f'https://graph.facebook.com/v18.0/act_{ad_account_id}/ads'
#time_range=f"{'since':'2023-10-11','until':'{dt.datetime.now().strftime('%Y-%m-%d')}'}"

params_ads = {
    'access_token': access_token,
    'fields': (
        'id,name,status,'
        'insights.time_increment(1){'
        'date_start,impressions,clicks,unique_clicks,cpc,cpm,ctr,spend,frequency,reach}'
    ),
    'summary': 'true',
    'time_range':time_range
}

response_ads = requests.get(url_ads, params=params_ads)
if response_ads.status_code == 200:
    ads_data = response_ads.json()['data']
    insights_list_ads = []
    for ad in ads_data:
        insights_ads = ad.get('insights', {}).get('data', [])
        for insight in insights_ads:
            insight_data_ads = {
                'date': insight['date_start'],
                'ad_id': ad['id'],
                'ad_name': ad['name'],
                'ad_status': ad['status'],
                **{key: insight[key] for key in insight if key != 'date_start'}
            }
            insights_list_ads.append(insight_data_ads)
    df_ads = pd.DataFrame(insights_list_ads)
    df_ads.set_index('date', inplace=True)
    print(df_ads)
else:
    print(f'Error: {response_ads.status_code}')
    print(response_ads.text)
```
